```python
import pandas as pd

# Зчитуємо CSV файл
df = pd.read_csv('D:/Download/NationalNames.csv')

# Фільтруємо чоловічі імена за 2010 рік
чоловічі_імена_2010 = df[(df['Year'] == 2010) & (df['Gender'] == 'M')]

# Знаходимо топ-5 чоловічих імен
топ_5_чоловічих_імен = чоловічі_імена_2010.sort_values('Count', ascending=False).head(5)

# Виводимо результати
print("5 найпопулярніших чоловічих імен у 2010 році:")
for index, row in топ_5_чоловічих_імен.iterrows():
    print(f"{row['Name']}: {row['Count']} разів")

```

```python
#Знайдіть найпопулярніше ім’я за результатами одного року (ім’я, для якого Count максимальне)
import pandas as pd

# Зчитуємо CSV файл
df = pd.read_csv('D:/Download/NationalNames.csv')
найпопулярніше = df.loc[df['Count'].idxmax()]
print(f"Найпопулярніше ім'я: {найпопулярніше['Name']}")
print(f"Рік: {найпопулярніше['Year']}")
print(f"Стать: {найпопулярніше['Gender']}")
print(f"Кількість: {найпопулярніше['Count']}")
```

```python
#Підрахуйте кількість записів, для яких Count - мінімальне у наборі.
import pandas as pd
# Зчитуємо CSV файл
df = pd.read_csv('D:/Download/NationalNames.csv')
мінімальний_count = df['Count'].min()
кількість_мінімальних = df[df['Count'] == мінімальний_count].shape[0]
print(f"Мінімальне значення Count: {мінімальний_count}")
print(f"Кількість записів з мінімальним Count: {кількість_мінімальних}")
```

```python
#Підрахуйте кількість унікальних імен у кожному році
import pandas as pd
# Зчитуємо CSV файл
df = pd.read_csv('D:/Download/NationalNames.csv')
унікальні_імена_по_роках = df.groupby('Year')['Name'].nunique().reset_index()
унікальні_імена_по_роках.columns = ['Рік', 'Кількість унікальних імен']
print(унікальні_імена_по_роках)
```

```python
#Знайдіть рік із найбільшою кількістю унікальних імен.
рік_максимум = унікальні_імена_по_роках.loc[унікальні_імена_по_роках['Кількість унікальних імен'].idxmax()]
print(f"Рік з найбільшою кількістю унікальних імен: {рік_максимум['Рік']}")
print(f"Кількість унікальних імен: {рік_максимум['Кількість унікальних імен']}")
```

```python
#Знайдіть найпопулярніше ім’я в році з найбільшою кількістю унікальних імен (тобто у 2008 році)
рік_макс_унікальних = рік_максимум['Рік']
дані_року = df[df['Year'] == рік_макс_унікальних]
найпопулярніше_імя = дані_року.loc[дані_року['Count'].idxmax()]
print(f"Найпопулярніше ім'я в {рік_макс_унікальних} році: {найпопулярніше_імя['Name']}")
print(f"Кількість: {найпопулярніше_імя['Count']}")
```

```python
#Підрахуйте кількість років, коли дівчаток народжувалось більше, ніж хлопчиків.
народження_по_роках = df.groupby(['Year', 'Gender'])['Count'].sum().unstack()
роки_більше_дівчаток = (народження_по_роках['F'] >народження_по_роках['M']).sum()
print(f"Кількість років, коли народилось більше дівчаток: {роки_більше_дівчаток}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Зчитуємо CSV файл
df = pd.read_csv('D:/Download/NationalNames.csv')

# Групуємо дані за роками та статтю і обчислюємо загальну кількість народжень
народження_по_роках = df.groupby(['Year', 'Gender'])['Count'].sum().unstack()

# Побудова графіку
plt.figure(figsize=(12, 6))
plt.plot(народження_по_роках.index, народження_по_роках['M'], label='Хлопчики', color='blue')
plt.plot(народження_по_роках.index, народження_по_роках['F'], label='Дівчатка', color='pink')
plt.xlabel('Рік')
plt.ylabel('Кількість народжень')
plt.title('Загальна кількість народжень хлопчиків та дівчаток на рік')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Підрахуйте кількість гендерно нейтральних імен (однакових для дівчат та хлопців)
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('D:/Download/NationalNames.csv')

імена_за_статтю = df.groupby('Name')['Gender'].nunique()
гендерно_нейтральні = імена_за_статтю[імена_за_статтю> 1]
кількість_нейтральних = len(гендерно_нейтральні)

print(f"Кількістьгендерно нейтральних імен: {кількість_нейтральних}")
```

```python
#Порахуйте, скільки разів хлопчиків називали Barbara
barbara_хлопчики = df[(df['Name'] == 'Barbara') & (df['Gender'] == 'M')]['Count'].sum()
print(f"Кількість хлопчиків з іменем Barbara: {barbara_хлопчики}")
```

```python
#Знати найпопулярніші гендерно-нейтральні імена (ті, що присутні кожного року)
роки = df['Year'].nunique()
нейтральні_імена = df.groupby('Name').filter(lambda x: x['Gender'].nunique() > 1 and x['Year'].nunique() == роки)
популярні_нейтральні = нейтральні_імена.groupby('Name')['Count'].sum().sort_values(ascending=False).head(10)

print("Найпопулярніші гендерно-нейтральні імена:")
print(популярні_нейтральні)

```

```python
#Побудувати графіки розподілення кількості імен John та Mary по роках без залежності до статі.
import matplotlib.pyplot as plt

john_mary = df[df['Name'].isin(['John', 'Mary'])].groupby(['Year', 'Name'])['Count'].sum().unstack()

plt.figure(figsize=(12, 6))
plt.plot(john_mary.index, john_mary['John'], label='John')
plt.plot(john_mary.index, john_mary['Mary'], label='Mary')
plt.xlabel('Рік')
plt.ylabel('Кількість')
plt.title('Розподілення кількості імен John та Mary по роках')
plt.legend()
plt.grid(True)
plt.show()
```
